##step 1 => uploading raw documents
Using
* colab file


```python
from google.colab import files

upload=files.upload()
file_name=list(upload.keys())[0]
```

*  Extracting text using Pymupdf
**Document loader**

```python
!pip install pymupdf
```

```python
from langchain_community.document_loaders import PyMuPDFLoader

content=PyMuPDFLoader(file_name).load()
print(content)
```

```python
content1=[doc.page_content for doc in content]
content="".join(content1)
print(content)
```

##step 2 => Converting raw text into Chunks
Using
* text splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

chunks=RecursiveCharacterTextSplitter(chunk_size=50,chunk_overlap=0).split_text(content)
print(chunks)
```

##step 3 => creating embeddings and store in vector stors

Using
* Chromadb
* googlegenerativeaiembeddings

Note
* vector stores => store only document

```python
from langchain_core.documents import Document
from google.colab import userdata
#
doc_chunks= [Document(page_content=text) for text in chunks]
print(doc_chunks)
```

Chromadb

```python
!pip install langchain_chroma Chromadb
```

```python
from chromadb import EmbeddingFunction
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_chroma import Chroma

llm=GoogleGenerativeAIEmbeddings(model="gemini-embedding-001",google_api_key=userdata.get('gemikey'))
vector_store=Chroma(
    embedding_function=llm,
    persist_directory="db",
    collection_name="crick_bot_web_search")
```

retriver method

```python
vector_store.add_documents(doc_chunks)
```

```python
retriever=vector_store.as_retriever(search_kwargs={"k":2})
query="formats of cricket"
response=retriever.invoke(query)
context=" "
for doc in response:
  context+=doc.page_content

print(context)
```

```python
!pip install langchain_tavily tavily
```

```python
from langchain_community.tools.tavily_search import TavilySearchResults
from google.colab import userdata

def web_search(query):
  search = TavilySearchResults(tavily_api_key=userdata.get('tavily'), include_raw_results=True, max_results=1)
  res_web=search.invoke(query)
  return res_web
```

LLM


```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
LLM=ChatGoogleGenerativeAI(model="gemini-2.5-flash",google_api_key=userdata.get('gemikey'))

prompt=PromptTemplate(
                      template=""" you are a cricket expert ,answer the following user query:{query}
                              based on context provided {context}
                              If enough information is not available in {context} for user query , response with "insufficient information".""",
                      input=["query","context"])

chain=prompt | LLM | StrOutputParser()

ress=chain.invoke({"query":query,"context":context})
print(ress)
```

```python
result=web_search(query)
print(result[0]["content"])
```

# **Chatbot**

```python
while True:
  #user giving input
  user_input=input("YOU:")
  if user_input.strip().lower() in ["exit","by","stop"]:
    print("exiting chatbot")
    break
  #gathering relevant content from chroma
  response=retriever.invoke(user_input)
  final_content=" "
  for doc in response:
    final_content+=doc.page_content
  #invokeing llm
  chatbot_response=chain.invoke({"query":user_input,"context":final_content})
  if chatbot_response == "insufficient information":
    chatbot_response=web_search(user_input)
    chatbot_response=chatbot_response[0]["content"]

  else:
    chatbot_response=chatbot_response

  print(f"chatbot_response:{chatbot_response}")
```
